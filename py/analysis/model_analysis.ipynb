# Market Data 1: Python

```python
fig = plt.figure(figsize=(15, 6), dpi=80)

#PnL given no realizing PnL
df = pd.read_csv('./../score_board.csv', sep=',')
model = df[df['Team'] == 'BaseModel1']

temp = (max(model['ProfitOrLoss']) + min(model['ProfitOrLoss'])) / 2
PnL = [(i - temp)/100 for i in model['ProfitOrLoss']]

plt.subplot(1, 2, 1)
plt.title('Profit Over Time')
plt.xlabel('Time')
plt.ylabel('Profit')
plt.plot(range(len(PnL)), PnL, color = 'black', label = 'Profit/Loss')

#     PnL_prev = [i/100 for i in prev['ProfitOrLoss']]
#     plt.subplot(1, 3, 1)
#     plt.plot(range(len(PnL_prev)), PnL_prev, color = 'red', label = 'Previous Model')
#     plt.legend()
# ------------------------------------------------------------------ #
# bar plot of our ETF position as time goes on.
plt.subplot(1, 2, 1)
scaled_heights = [i * (max(PnL) / 100) for i in df['EtfPosition']]
# plt.title('Etf Position Over Time')
# plt.xlabel('Time')
# plt.ylabel('Position')
plt.bar(range(len(scaled_heights)), scaled_heights, width = 1, color = 'black', alpha = 0.3, label = 'Etf Position')
plt.legend()
 
# ------------------------------------------------------------------ #
# etf price of the thang we traded
# temp = min(df['EtfPrice'])
# df['EtfPrice'] = [i - temp for i in df['EtfPrice']]
temp = (min(df['EtfPrice'].dropna()) + max(df['EtfPrice'].dropna())) / 2
etf_price_of_simulation = pd.DataFrame([i/100 for i in df['EtfPrice'].apply(lambda row: row - temp)])

plt.subplot(1, 2, 2)
plt.title('Etf Price of Stock We Just Traded')
plt.xlabel('Time')
plt.ylabel('Price')
plt.plot(range(len(etf_price_of_simulation)), etf_price_of_simulation, color = 'black', label = 'Etf Price')

plt.subplot(1, 2, 2)
scaled_heights = [i * (max(etf_price_of_simulation.dropna()[0]) / 100) for i in df['EtfPosition']]
# plt.title('Etf Position Over Time')
# plt.xlabel('Time')
# plt.ylabel('Position')
plt.bar(range(len(scaled_heights)), scaled_heights, width = 1, color = 'black', alpha = 0.3, label = 'Etf Position')
plt.legend()

# check Sharpe ratio
temp = list(df.ProfitOrLoss)
print(f'SHARPE RATIO ON SAMPLE DATA 1 IS {np.mean(temp) / np.std(temp)}')
```

```python
fig = plt.figure(figsize=(15, 15), dpi=80)

#PnL given no realizing PnL
df = pd.read_csv('./../round2/match125_score_board.csv', sep=',')
# df = pd.read_csv('./../score_board.csv', sep=',')
models = dict()

for name in set(df['Team']):
    models[name] = df[df['Team'] == name]

for name, model in models.items():
    print(name)
    PnL = [i/100 for i in model['ProfitOrLoss']]
    plt.subplot(2, 1, 1)
    plt.title('Profit Over Time of Simulation')
    plt.xlabel('Time')
    plt.ylabel('Profit')
    plt.plot(range(len(PnL)), PnL, label = name)
    
plt.legend()

plt.subplot(2, 1, 2)
temp = models['LiquidBears_2185']['EtfPrice']
plt.plot(range(len(temp)), temp, label = 'EtfPrice')
plt.legend()

# ---- SHARPE RATIOS FOR EACH TEAM / MODEL WITHIN THE SIMULATION
sharpe_ratios = dict()

match = pd.read_csv('./../round2/match125_score_board.csv', sep=',')
# match = pd.read_csv('./../score_board.csv', sep=',')

for model_name in set(df['Team']):
    temp = list(match[match['Team'] == model_name]['ProfitOrLoss'])
    sharpe_ratios[model_name] = (np.mean(temp) / np.std(temp))
    
for one, two in sharpe_ratios.items():
    print(one, ': ', two, sep ='', end='\n')
```
